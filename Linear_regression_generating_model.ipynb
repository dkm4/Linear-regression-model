# Import new libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Prepare some dataset

```python
print(range(100))
print([i for i in range(100)])
x = [i for i in range(100)]
print(x)
my_data = pd.DataFrame(x)
my_data.rename(columns={0:"x"}, inplace=True)
my_data['y'] = my_data['x']*5
my_data.to_csv("some_sample_dataset.csv", index=None)
```

# Visualize the file

```python
my_data = pd.read_csv("some_sample_dataset.csv")
my_data
```

#Plot the data


```python
plt.scatter(my_data['x'].values.tolist(), my_data['y'].values.tolist(), color='blue', label="Full Dataset", s=1)
plt.legend()
```

# Split the dataset into 2 parts:Seen/unseen data

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(my_data, test_size=0.25, random_state=42)
print(len(train_df))
print(len(test_df))
plt.scatter(train_df['x'].values.tolist(), train_df['y'].values.tolist(), color='blue', label="Training Dataset", s=1)
plt.scatter(test_df['x'].values.tolist(), test_df['y'].values.tolist(), color='red', label="Test Dataset", s=1)
plt.legend()
```

# Build a model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
print(model)
```

# Train the model

```python
model.fit(train_df[['x']], train_df['y'])
print("training is completed")
```

# Visualize the model

```python
print(model.coef_)
print(model.intercept_)
```

# Generate output for new examples

```python
test_df['predicted_y'] = model.predict(test_df[['x']])
test_df
```
